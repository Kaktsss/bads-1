For example, say you want to incorporate the famous *Cobbâ€“Douglas* function in your notebook, you can do so as follows: 

$$ Y \left(L,K \right) =AL^\beta K^\alpha$$

where:
- $Y =$ total production (the real value of all goods produced in a year or 365.25 days)
- $L =$ labour input (person-hours worked in a year or 365.25 days)
- $K =$ capital input (a measure of all machinery, equipment, and buildings; the value of capital input divided by the price of capital)
- $A =$ [total factor productivity](https://en.wikipedia.org/wiki/Total_factor_productivity)
- $0 \le \alpha \le 1$ and $0 \le \beta \le 1$ are the output elasticities of capital and labor, respectively. These values are constants determined by available technology.

It contains a bunch of statements to plot the [Rosenbrock function](https://en.wikipedia.org/wiki/Rosenbrock_function). It is fine if you have not heard about this function before. The only reason to consider it here is that its shape look somewhat nice, which is the reason for the function also being known as the *banana function*. 

To plot the function, simply run the next code cell. It is safe to ignore the code, which might look overwhelming at first glance.

````python
# Python code to create a plot of the Rosenbrock function
import numpy as np

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100*(y - x**2)**2

# Create a mesh grid of x and y values
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)

# Calculate the z values
Z = rosenbrock(X, Y)

# Plot the 3D surface
fig = plt.figure(figsize=[12,12])
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.title("Rosenbrock function")
plt.show()
````

Just in case you are interested in the mathematical underpinnings of the *Rosenbrock function*, here is how it is defined:

$$f(x, y) = (1 - x)^2 + 100(y - x^2)^2$$

````python
# Empty code cell to which you can copy-paste ChatGPT output
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# --- 1. Define Parameters ---
# A: Total Factor Productivity (Scale factor)
A = 1.0   
# alpha: Output elasticity of capital (usually < 1)
alpha = 0.5 
# beta: Output elasticity of labor (usually < 1)
beta = 0.5  

# --- 2. Generate Data Grid ---
# Create a range of values for Capital (K) and Labor (L)
# linspace creates 100 evenly spaced points between 0 and 10
k_values = np.linspace(0, 10, 100)
l_values = np.linspace(0, 10, 100)

# Create a coordinate grid (mesh) for 3D plotting
# K and L become 2D arrays representing the floor of the plot
K, L = np.meshgrid(k_values, l_values)

# --- 3. Calculate Output (Y) ---
# Apply the Cobb-Douglas formula to the entire grid
Y = A * (K ** alpha) * (L ** beta)

# --- 4. Create the 3D Plot ---
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

# Plot the surface
# cmap='viridis' adds color depth based on the height (Output)
surf = ax.plot_surface(K, L, Y, cmap='viridis', edgecolor='none', alpha=0.9)

# --- 5. Labels and Aesthetics ---
ax.set_title(f'Cobb-Douglas Production Function\n($\\alpha={alpha}, \\beta={beta}, A={A}$)')
ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labor (L)')
ax.set_zlabel('Output (Y)')

# Adjust the viewing angle for a better perspective (elevation, azimuth)
ax.view_init(elev=30, azim=45)

# Add a color bar to indicate Output levels
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Output Level')

# Show the plot
plt.show()
````
